```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rainfall = [17,10,10,10,26,59]
month = ['jan','feb','mar','apr','may','jun']
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')#theme
```

```python
plt.bar(month,rainfall)
```

```python
plt.pie(
    rainfall,
    labels = month,
    autopct = '%1.1f%%',
    
)
plt.show()
```

```python
plt.pie(
    rainfall,
    labels = month,
    autopct = '%.1f%%',
    wedgeprops={'width':0.4,'edgecolor':'#6666ff','linewidth':2},
    pctdistance=.8
    
)
plt.show()
```

```python
plt.boxplot(rainfall,
            vert=False,
            showmeans=True,
            meanline=True,
            meanprops={'color':'red'}
            )
plt.show()
```
