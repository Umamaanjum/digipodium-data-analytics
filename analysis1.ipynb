```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
marks = pd.read_excel('marksheet_faked.xlsx')
marks
```

```python
marks.set_index('First name',inplace=True)
marks
```

```python
marks[['ECO','GFX','DM','SE','DBMS']].plot(kind='bar',figsize=(10, 5), title='results')
```

```python
marks['ECO'] +5#selecting a columns
```

```python
marks[['DBMS','ECO']] +3
```

```python
marks['total']=marks['DBMS']+marks['GFX']+marks['ECO']+marks['DM']+marks['SE']
marks
```

```python
marks['total'].plot(kind='bar',figsize=(10, 5), title='results')
```

```python
#sorting
marks.sort_values(by='total', ascending=False , inplace=True)
```

```python
marks['percentage']= marks['total'] / 5
```

```python
marks.sort_values(by=['total','percentage'],ascending=False, inplace=True)
```

```python
fig,ax = plt.subplots(figsize=(15,5),ncols=2)
marks['total'].plot(kind='bar',ax=ax[0])
cols = ['DBMS','DM','SE','ECO','GFX']
marks[cols].plot(kind='kde', ax=ax[1])
plt.show()
```
